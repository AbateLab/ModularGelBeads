```python
import numpy as np
import matplotlib.pyplot as plt
import barcode_methods as bcm 
import os
import Levenshtein
```

```python
if not 'path' in globals():
    path = os.getcwd()
print('path: ' + path)

file1 = os.path.join(path, 'fastq/tube4_S1_L001_R1_001.fastq.gz')
file2 = os.path.join(path, 'fastq/tube4_S1_L001_R2_001.fastq.gz')

```

```python
# all three barcode blocks are derived from the same 8nt edit distance 4 set
with open(os.path.join(path, 'gelbarcodes'), 'r') as f:
    bc_list = [line.strip().split(':')[1] for line in f][1:]

bc_list_comp = []
for i in bc_list:
    bc_list_comp.append(bcm.rev_comp(i))
```

```python
# create the sequencing experiment instance
temp_dat = bcm.seq_experiment([bc_list, bc_list, bc_list], bcerror_tolerance=1)

# define the common barcode structure for the beads
barcode_seq = 'NNNNNNNNAACCNNNNNNNNACAGNNNNNNNN'

# start collecting barcode sequences from reads
temp_dat.process_barcodes(file1, file2, barcode_seq,  verbose=True, rev_complement_s2=True)
```

```python
cells = []
for i in temp_dat.bc_groups.values():
    cells.append(len(i))
```

```python
try:
    os.makedirs("qc_output")
except OSError:
    pass

plt.plot(np.sort(cells)[::-1])
plt.yscale('log')
plt.xscale('log')
plt.xlabel('barcode groups')
plt.ylabel('reads')
plt.savefig('qc_output/kneeplot.png', dpi=600)
```

```python
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import sys, collections
```

```python
#load human transcript

tr2g = {}
trlist = []
with open('/drive3/Cyrille/Gelbeads_RNA/homo_sapiens/transcripts_to_genes.txt') as f:
    for line in f:
        l = line.split()
        tr2g[l[0]] = l[1]
        trlist.append(l[0])

genes = list(set(tr2g[t] for t in tr2g))

# load equivalence classes
ecs = {}
with open('/drive3/Cyrille/Gelbeads_RNA/kallisto_out/matrix.ec') as f:
    for line in f:
        l = line.split()
        ec = int(l[0])
        trs = [int(x) for x in l[1].split(',')]
        ecs[ec] = trs
        
def ec2g(ec):
    if ec in ecs:
        return list(set(tr2g[trlist[t]] for t in ecs[ec]))        
    else:
        return []
```

```python
cell_gene = collections.defaultdict(lambda: collections.defaultdict(float))
pbar=None
pumi=None
with open('/drive3/Cyrille/Gelbeads_RNA/kallisto_out/output.sort.txt') as f:
    gs = set()
    for line in f:
        l = line.split()
        barcode,umi,ec,count = line.split()
        ec = int(ec)
        
        if barcode == pbar:
            # same barcode
            if umi == pumi:
                # same UMI, let's update with intersection of genelist
                gl = ec2g(ec)
                gs.intersection_update(gl)
            else:
                # new UMI, process the previous gene set
                for g in gs:
                    cell_gene[barcode][g] += 1.0/len(gs)
                # record new umi, reset gene set
                pumi = umi
                gs = set(ec2g(ec))
        else:
            # work with previous gene list
            for g in gs:
                cell_gene[pbar][g] += 1.0/len(gs)
            
            if sum(cell_gene[pbar][g] for g in cell_gene[pbar]) < 10:
                del cell_gene[pbar]
            
            pbar = barcode
            pumi = umi
            
            gs = set(ec2g(ec))
    #remember the last gene
    for g in gs:
        cell_gene[pbar][g] += 1.0/len(gs)
        
    if sum(cell_gene[pbar][g] for g in cell_gene[pbar]) < 10:
        del cell_gene[pbar]
```
